```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('/home/moses/Desktop/uber/uber.csv')
```

```python
#checking for missing values
print(df.isnull().sum())
print(df.duplicated().sum())
```

```python
df=df.dropna(subset=['dropoff_longitude','dropoff_latitude'])
print(df)
```

```python
#creating a scatter plot of the pick up and the drop off locations
plt.scatter(df['pickup_longitude'], df['pickup_latitude'], s=2, alpha=0.9, label='pickup')
plt.scatter(df['dropoff_longitude'], df['dropoff_latitude'], s=2, alpha=0.9, label='dropoff')
# Add labels and legend
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Pickup and Dropoff Locations')
plt.legend()

# Show the plot
plt.show()
```

```python
#plotting histogram for fare_amount
plt.hist(df['fare_amount'], bins=50)
plt.title('Histogram of fare amount')
plt.xlabel('fare_amount($)')
plt.ylabel('frequency')
plt.show()
```

```python
# Plot histogram for typical fare rides
typical_fare_rides = df[df['fare_amount'] <= 50]
plt.hist(typical_fare_rides['fare_amount'], bins=50)
plt.title('Histogram of Typical Fare Rides')
plt.xlabel('Fare Amount ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
#create a boxplot for the passenger count
plt.boxplot(df['passenger_count'])
#title of the box
plt.title('Number of Passengers')
plt.ylabel('passenger count')
plt.show()
```

```python
df = pd.read_csv('/home/moses/Desktop/uber/uber.csv', parse_dates=['pickup_datetime'])
df['count'] = 1

# set the datetime column as the index
df.set_index('pickup_datetime', inplace=True)

#data to hourly intervals
hourly = df.resample('H').sum()

# time series data
plt.plot(hourly.index, hourly['count'])
plt.xlabel('Date')
plt.ylabel('Ride Count')
plt.title('Hourly Ride Counts')
plt.show()
```
